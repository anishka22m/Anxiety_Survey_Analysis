```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Load the dataset (assuming it's already loaded as a DataFrame named 'df')
# Replace 'file_path' with the path to your CSV if needed.
# df = pd.read_csv('file_path')

# Define the features and target variables
X = df.drop(columns=['Anxiety_Level', 'Stress_Level'])

# Targets for analysis
y_anxiety = df['Anxiety_Level']
y_stress = df['Stress_Level']

# Split the data into training and testing sets
X_train, X_test, y_anxiety_train, y_anxiety_test = train_test_split(X, y_anxiety, test_size=0.3, random_state=42)
X_train, X_test, y_stress_train, y_stress_test = train_test_split(X, y_stress, test_size=0.3, random_state=42)

# Initialize Random Forest Regressor
rf_anxiety = RandomForestRegressor(n_estimators=100, random_state=42)
rf_stress = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the models
rf_anxiety.fit(X_train, y_anxiety_train)
rf_stress.fit(X_train, y_stress_train)

# Get feature importances
importance_anxiety = rf_anxiety.feature_importances_
importance_stress = rf_stress.feature_importances_

# Create a DataFrame for feature importances
feature_importance_df_anxiety = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importance_anxiety
}).sort_values(by='Importance', ascending=False)

feature_importance_df_stress = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importance_stress
}).sort_values(by='Importance', ascending=False)

# Plot the feature importances for Anxiety_Level
plt.figure(figsize=(12, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df_anxiety.head(10))
plt.title('Top 10 Important Features for Predicting Anxiety Level')
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.show()

# Plot the feature importances for Stress_Level
plt.figure(figsize=(12, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df_stress.head(10))
plt.title('Top 10 Important Features for Predicting Stress Level')
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc


# Select the important features identified from the feature importance plot for both Anxiety and Stress
important_features_anxiety = ['Anxiety_Symptoms_Frequency', 'Control_Loss_Days', 'Peer_support', 
                              'Relaxation_Techniques_Frequency', 'Workload_Stress_Level', 
                              'Overwhelmed_Frequency', 'Sleep_Quality', 'Physical_Activity_Frequency']

important_features_stress = ['Anxiety_Symptoms_Frequency', 'Overwhelmed_Frequency', 
                             'Relaxation_Techniques_Frequency', 'Workload_Stress_Level', 
                             'Control_Loss_Days', 'Peer_support', 'Sleep_Quality', 
                             'Physical_Activity_Frequency', 'Age', 'Hobbies_Frequency']

# Ensure 'Sleep_Quality' and 'Physical_Activity_Frequency' are included in the features if not in the top
if 'Sleep_Quality' not in important_features_anxiety:
    important_features_anxiety.append('Sleep_Quality')

if 'Physical_Activity_Frequency' not in important_features_anxiety:
    important_features_anxiety.append('Physical_Activity_Frequency')

if 'Sleep_Quality' not in important_features_stress:
    important_features_stress.append('Sleep_Quality')

if 'Physical_Activity_Frequency' not in important_features_stress:
    important_features_stress.append('Physical_Activity_Frequency')

# Prepare the data
X_anxiety = df[important_features_anxiety]
y_anxiety = df['Anxiety_Level']

X_stress = df[important_features_stress]
y_stress = df['Stress_Level']

# Split the data into training and testing sets
X_anxiety_train, X_anxiety_test, y_anxiety_train, y_anxiety_test = train_test_split(X_anxiety, y_anxiety, test_size=0.3, random_state=42)
X_stress_train, X_stress_test, y_stress_train, y_stress_test = train_test_split(X_stress, y_stress, test_size=0.3, random_state=42)

# Create and train logistic regression models
log_reg_anxiety = LogisticRegression(max_iter=1000)
log_reg_anxiety.fit(X_anxiety_train, y_anxiety_train)

log_reg_stress = LogisticRegression(max_iter=1000)
log_reg_stress.fit(X_stress_train, y_stress_train)

# Predict and save results to a CSV
df['Predicted_Anxiety_Level'] = log_reg_anxiety.predict(X_anxiety)
df['Predicted_Stress_Level'] = log_reg_stress.predict(X_stress)
df.to_csv('predicted_results.csv', index=False)

# Print evaluation metrics
print("Anxiety Level Classification Report:")
print(classification_report(y_anxiety_test, log_reg_anxiety.predict(X_anxiety_test)))

print("Stress Level Classification Report:")
print(classification_report(y_stress_test, log_reg_stress.predict(X_stress_test)))

# Plot confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Confusion matrix for Anxiety Level
sns.heatmap(confusion_matrix(y_anxiety_test, log_reg_anxiety.predict(X_anxiety_test)), annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title('Confusion Matrix for Anxiety Level')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

# Confusion matrix for Stress Level
sns.heatmap(confusion_matrix(y_stress_test, log_reg_stress.predict(X_stress_test)), annot=True, fmt='d', cmap='Blues', ax=axes[1])
axes[1].set_title('Confusion Matrix for Stress Level')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')

plt.tight_layout()
plt.show()

# Plot ROC Curves
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# ROC for Anxiety Level
y_anxiety_prob = log_reg_anxiety.predict_proba(X_anxiety_test)[:, 1]
fpr_anxiety, tpr_anxiety, _ = roc_curve(y_anxiety_test, y_anxiety_prob, pos_label=log_reg_anxiety.classes_[1])
roc_auc_anxiety = auc(fpr_anxiety, tpr_anxiety)
ax[0].plot(fpr_anxiety, tpr_anxiety, label=f'Anxiety Level (AUC = {roc_auc_anxiety:.2f})')
ax[0].plot([0, 1], [0, 1], 'k--')
ax[0].set_title('ROC Curve for Anxiety Level')
ax[0].set_xlabel('False Positive Rate')
ax[0].set_ylabel('True Positive Rate')
ax[0].legend()

# ROC for Stress Level
y_stress_prob = log_reg_stress.predict_proba(X_stress_test)[:, 1]
fpr_stress, tpr_stress, _ = roc_curve(y_stress_test, y_stress_prob, pos_label=log_reg_stress.classes_[1])
roc_auc_stress = auc(fpr_stress, tpr_stress)
ax[1].plot(fpr_stress, tpr_stress, label=f'Stress Level (AUC = {roc_auc_stress:.2f})')
ax[1].plot([0, 1], [0, 1], 'k--')
ax[1].set_title('ROC Curve for Stress Level')
ax[1].set_xlabel('False Positive Rate')
ax[1].set_ylabel('True Positive Rate')
ax[1].legend()

plt.tight_layout()
plt.show()
```
